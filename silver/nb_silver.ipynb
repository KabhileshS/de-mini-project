```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql import functions as F
spark = SparkSession.builder.getOrCreate()
def to_correct_column(name: str) -> str:
    a=""
    for i in range(0,len(name)):
      if i==0:
        a+=name[i]
        continue
      if name[i-1]==name[i]=='_':
        continue
      else:
        a+=name[i]
    l=list(map(str,a.split('_')))
    b=""
    for i in l:
      if len(i)==1:
        b+=i
      else:
        b+=i+"_"
    if(b[-1]=='_'):
      b=b[:len(b)-1:]
    name=b
    return name
```

```python
df = spark.read.table("bronze.data_dictionary")
df = df.toDF(*[to_correct_column(c) for c in df.columns])

df= df.withColumn("table", col("table").cast("string"))
df= df.withColumn("field", col("field").cast("string"))
df= df.withColumn("description", col("description").cast("string"))

df_cleaned = df.fillna({
    "table": "Unknown",     # Replace null driver_name
    "field": "Unknown",        # Replace null distance_miles
    "description": "Unknown"         # Replace null delay_in_minutes
})
df_cleaned=df_cleaned.dropDuplicates()
# df_cleaned.show()
# display(df_cleaned)
df_cleaned.write.format("delta").mode("overwrite").saveAsTable("silver.data_dictionary")
```

```python
from pyspark.sql.functions import regexp_replace
df = spark.read.table("bronze.customers")

df = df.toDF(*[to_correct_column(c) for c in df.columns])

df= df.withColumn("customer_key", col("customer_key").cast("int"))
df= df.withColumn("gender", col("gender").cast("string"))
df= df.withColumn("name", col("name").cast("string"))
df= df.withColumn("city", col("city").cast("string"))
df= df.withColumn("state_code", col("state_code").cast("string"))
df= df.withColumn("state", col("state").cast("string"))
df= df.withColumn("zip_code", col("zip_code").cast("string"))
df= df.withColumn("country", col("country").cast("string"))
df= df.withColumn("continent", col("continent").cast("string"))

df=df.withColumn('birthday',
regexp_replace(col('birthday'), '[-.]', '/'))
df=df.withColumn('birthday',
coalesce(
    try_to_date(col('birthday'), 'MM/dd/yyyy'),
    try_to_date(col('birthday'), 'M/dd/yyyy'),
    try_to_date(col('birthday'), 'MM/d/yyyy'),
    try_to_date(col('birthday'), 'M/d/yyyy'),
    try_to_date(col('birthday'), 'yyyy/MM/dd'),
    try_to_date(col('birthday'), 'dd/MM/yyyy'),
    try_to_date(col('birthday'), 'yyyy-MM-dd')
)
)
# df=df.withColumn("birthday",
#         F.coalesce(
#             F.to_date(F.col("birthday"), "MM/dd/yyyy"),
#             F.to_date(F.col("birthday"), "M/d/yyyy"),
#             F.to_date(F.col("birthday"), "MM/dd/yyyy"),
#             F.to_date(F.col("birthday"), "M/dd/yyyy"),
#             F.to_date(F.col("birthday"), "yyyy/MM/dd"),
#             F.to_date(F.col("birthday"), "dd/MM/yyyy"),
#             F.to_date(F.col("birthday"), "yyyy/MM/dd" ))
#         )

df_cleaned = df.fillna({
    "gender": "Unknown",     
    "name": "Unknown",        
    "state_code": "Unknown"  ,       
    "city": "Unknown",    
    "state": "Unknown",        
    "continent": "Unknown",
    "country": "Unknown"  , 
    "zip_code": "Unknown",
})

df_cleaned=df_cleaned.dropDuplicates()
df_cleaned.write.format("delta").option("overwriteSchema", "true").mode("overwrite").saveAsTable("silver.customers")

# df_cleaned.show()
# df_cleaned.printSchema()
# df.show()
```
